### __Magic commands__
<p> Magic commands in Python are special functions used to modify the behavior of a 
code cell or a line in at the time of execution. Magic commands allow users to run
special non-python code in a cell, e.g. %%bash for Bash cell commands, %%html for html code.
%matplotlib inline displays the figure in a notebook and not in a separate window when
plt.show() command is used.</>

### **Bar charts, count plots, and histograms**

<p> Bar plots and count plots display categorical variables as different bars on the X-axis. However, bar 
plots are mostly used to display means of quantitative variables while count plots display frequencies
of categorical variables</>
<p> Histograms display the frequency distribution of a quantitative variable, such as the ages of 
of participants. </>


```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [9,8,7,6,5,4,3,2,1]
```

```python
# Matplotlib line
plt.figure(figsize=(5,3))

plt.plot(x, y)
plt.title("My Line Plot")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
# Seaborn line plot
plt.figure(figsize = (5,3))

sns.lineplot(x=x, y=y)

plt.show()
```

```python
# Pie chart
gender = ('male', 'female')
proportion = (30, 70)
```

```python
plt.figure(figsize=(5, 3))

plt.pie(proportion, labels= gender, startangle = 90, autopct='%1.1f%%')
plt.show()
```

```python
# SNS countplot
sns.countplot(x= proportion, hue=gender)

plt.show()
```

```python
# Bar chart matplotlib
plt.figure(figsize=(5,3))
plt.bar(x, y)
plt.show()
```

```python
# Seaborn bar chart
sns.barplot(x=x, y=y, hue = x, palette='tab10')
plt.show()
```

```python
# Box plot
plt.boxplot(x, vert = False, notch = False)
plt.show()
```

```python
sns.boxplot(x)
plt.show()
```

```python
# Import plotly
import plotly.express as px
import pandas as pd
```

```python
# Create dataframe
df = pd.DataFrame({
    "x": x,
    "y": y
})
```

```python
# Line plot

px.line(df, x="x", y="x", width= 500, height = 400, title= "Lineplot")
```

```python
# Bar plots
px.bar(df, x='x', y='y', width = 500, height = 400, color ="y")
```

```python
# Histogram
px.histogram(df, x= "x", nbins = 3)
```

```python
# Pie chart

df_pie = pd.DataFrame({
    "proportions":proportion,
    "gender": gender
})

px.pie(df_pie, values = 'proportions', names='gender', width = 500, height = 400)
```

```python
# Scatterplot
px.scatter(df, x="x", y="y", width = 500, height= 400)
```
