```python
# Load dataset
inflation_data = pd.read_excel(r'C:\Users\user\Documents\LuxDevDS&AI\Kenya_inflation_wb.xlsx',\
                               index_col=('year'), parse_dates=['year'])
```

```python
# Visualize
plt.figure(figsize=(8,6))
plt.plot(inflation_data)
plt.title("Kenya Inflation 1960-2023")
plt.ylabel("Annual Inflation %")
plt.show()
```

```python
# Create Lag features
inflation_data['lag1'] = inflation_data['Inflation(annual %)'].shift(1)

# Calculate growth rates
inflation_data['growth_rate'] = round((inflation_data['Inflation(annual %)'] - 
                                 inflation_data['lag1']) / inflation_data['Inflation(annual %)']*100, 2)
```
