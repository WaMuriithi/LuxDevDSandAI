### Understand the difference between loading (json) and reading (CSV ) 
<p>
Handling a JSON file in Python is referred to as loading because JSON files contain data formatted in KEY:VALUE
pairs and may be nested- sometime on several levels- thus the data must be parsed before being read into memory.</p> 
<p> On the other hand, CSV is a flat file format recording data in lines which can be read directly into memory
</p>

### The difference between writing (CSV) and dumping (json)
<p>
As explained above JSON files are not flat, thus converting a a dataset in flat structure to JSON requires
the program to create key:value pairs and nesting structure hence the process is called dumping, as opposed
to writing a flat file such as CSV or txt
</p>
